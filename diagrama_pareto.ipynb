 Diagrama de Pareto


Um diagrama de pareto é uma ferramenta da qualidade que tem como objetivo identificar quais probemas são
responsáveis pelas maiores causas de problemas e quais apresentam maiores percentuais de redução de problemas 

Esse diagrama pode-se dizer, que é formado por duas partes, uma representa a freqencia das causas 
sendo essa parte formada por gráfico de barras e outra formada por uma reta que representa o quanto
dos problemas seriam resolvidos se a determinada causa fosse inibida.

O diagrama de pareto é uma ferramenta desenvolvida por Joseph Juran com base no Princípio de Pareto
onde se aplica a regra 80-20 (20% dos Clientes são Responsáveis por 80% do faturamento)

```python
# Importação das Biblotecas Necessárias para se trabalhar com os dados
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import math as mt
import matplotlib.gridspec as gridspec
```

```python
# Fazendo a Importação dos Dados de arquivos para dentro do Python

"""
para se importar dados no Pandas, é necessário que o arquivo esteja no formato suportado, para fins de 
padronização, usaremos o arquivo .csv para a importação

também para fins de facilitação, o arquivo deve está no mesmo diretório que o algoritmo está salvo

"""
# usaremos a função pd.read_csv("nome do arquivo para podermos importar")

```

```python
df = pd.read_csv("pareto.csv")


"""
 Nota: Células em arquivos csv precisam está separados por virgula, isso fará com que 
o Pandas/Python enterpretem corretamente como uma unidade do Dataframe

"""


percentual = (df["FREQUENCIA"]/df["FREQUENCIA"].sum()*100) # Ele vai calcular os valores de cada celula e dividir por 100
df["PERCENTUAL"]=percentual
df["FREQUENCIA ACUMULADA"] = df["PERCENTUAL"].cumsum()
df.drop([10]) # Remover a ultima linha


# Chamando o data frame, o Jupyter irá mostrar. Se estiver usando o Vscode, poderá abrir uma aba só com a tabela clicando
# em Variables e no link que se refere a tabela
```

```python
# Criando Gráficos para A tabela

"""
Umas das coisas mais legais do Python é que ele é como um canivete suíço, ou seja, tem ferramenta para
tudo. O Python possui ferramentas funções imbutidas para geração de figuras mas vamos trabalhar
com aquelas que são mais usadas pelos Cientistas de Dados

Matplotlib e o Seaborn



"""
```

```python
# Criando um Gráifico de Barras no Matplotlib

"""
Queremos um gráfico de barras para representar as frequências dos problemas e então um gráfico de
pontos que represente o percentual de resolução dos problemas 
"""

plt.figure(num = 1, figsize = (10,10))
plt.axis()
plt.subplot(3,3,2)
plt.xlabel("Eventos")
plt.ylabel("Percentual de Causa")
plt.title("Diagrama de Pareto")
plt.show()

"""
Nesse momento, criamos as entidades básicas de um gráfico. A região de Plotagem (Aqui chamada
de figure), as regiões que serão os eixos x e y (axis), os titulos dos eixos x e y (y/x label)
e a função que faz com que o gráfico seja gerado (Obrigatóriamente, os dois primeiros e o ultimo
devem existir)

É possivel simplificar o tamanho do código para que o gráfico seja plotado, mas por enquanto vamos
manter esse modelo para entendermos a lógica da geração do gráfico

Quando queremos trabalhar com gráficos lineares usamos a função pyplot, existem outras funções 
para outros tipos de gráficos. Podemos trabalhar também usando o pyplot.plot e pyplot.bar para
trabalharmos com gráficos lineares e gráficos de barras respectivamentes


"""

```

```python
# Fazendo a Plotagem de Pontos num Gráfico usando um dataframe do Pandas

"""

Vamos Esquecer o nosso problema por enquanto e queremos usar uma tabela com valores de vendas
relacionando os meses os valores das vendas nos meses


        MESES       VENDAS
    JANEIRO         R$1500,00
    FEVEREIRO       R$1200,00
    MARÇO           R$1600,00
    ABRIL           R$1300,00
    MAIO            R$1900,00

No Excel, seria bem simples, selecionar as duas colunas e gerar gráfico, mas no Pandas?

bom , o primeiro passo é criar um DataFrame("a planilha do pandas") e organizar os dados. Podemos 
usar um arranjo do Numpy para gerar esse dataframe mas vamos usar o numpy apenas para cálculo
juntamente com os modulos matemáticos Built - In (cmath, math)

Como sabemos o Pandas trata um arranjo unidimensional (ou coluna) como o tipo "SERIES" e um
arranjo multidimensional como o "DataFrame"

"""

#Para gerar um dataframe no Pandas, usamos a Função pandas(ou pd).DataFrame(Iniciais Maúsculas)()
"""
 Um dos meios de entradas é usando um Dicionário e uma Lista, O Python e o Pandas associará a Key do
discionário como o título e os valores da lista como os valores da coluna. Lembre-se indexação no
Python começa com 0 e o Pandas associará cada valor em cada coluna com base na lista

"""
dt_vendas = pd.DataFrame({"Meses":["Janeiro","Fevereiro","Março","Abril", "Maio"],"Valores":[1500,1200,1600,1300,1900]})

print(dt_vendas) # Usando a Função Print para voltar o texto
dt_vendas 

# Chamando a Tabela pela variável, temos ela sendo carregada no formato HTML o que 
# Visualmente fica mais elegante
```

```python
# Plotando um Gráfico Linear com Matplotlib

# Vamos associar os valores da primeira coluna com x e os da segunda com y

x = dt_vendas["Meses"] # Chamando pela Chave temos a extração dos valores da Coluna Meses
y = dt_vendas["Valores"]

plt.figure(num = 1, figsize = (10,10))
plt.plot(x,y,"gs-",) # A função plot irá plotar x e y e gerará uma linha reta com pontos na interceções
# Para mais configurações visite https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot
plt.xlabel("Vendas")
plt.ylabel("Valores")
plt.title("Vendas Mensais")
plt.show

```

```python
# Fazendo Personalizações no Gráfico

# - Colocando Uma Legenda

"""
Método Automático:

"""
plt.figure(num = 1, figsize = (10,10))
plt.plot(x,y,"gs-",)
plt.xlabel("Vendas")
plt.ylabel("Valores")
plt.legend(["Valores"],loc = "upper center",title = "Legenda") # Coloque o nome da legenda numa lista e a localização usando o método loc
plt.title("Vendas Mensais")
plt.show
```

```python
#Alterando o Fundo do Gráfico


# Algumas referencias para você personalizar o fundo do seu gráfico
# https://matplotlib.org/stable/gallery/style_sheets/dark_background.html
# https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
# https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html

plt.style.use('dark_background')
plt.figure(num = 1, figsize = (10,10))
plt.plot(x,y,"gs-",)
plt.xlabel("Vendas")
plt.ylabel("Valores")
plt.legend(["Valores"],loc = 5) # Coloque o nome da legenda numa lista e a localização usando o método loc
plt.title("Vendas Mensais")
plt.show
```

#Plotando um Gráfico de Barras


"""

Vamos supor que agora temos que fazer gráficos de barras para mostrar quantas vendas foram
feitas em cada mês


    MESES       VENDAS
    JANEIRO         20
    FEVEREIRO       30
    MARÇO           45
    ABRIL           12
    MAIO            53

vamos usar o método bar para plotar um gráfico de barras
"""


```python
dt_barras = ({"Janeiro":20,"Fevereiro":30,"Março":45,"Abril":12,"Maio":53})

#Criando um Dicionário para Criar uma referencia de Meses e Frequencia

meses = list(dt_barras.keys())
frequencia = list(dt_barras.values())
# a função bar plota gráficos verticais, sendo o primeiro argumento os nomes dos gráficos e o segundo os valores

plt.figure(num = 1, figsize = (10,10))
plt.style.use('ggplot')
plt.bar(meses,frequencia,color = "red")
plt.show
#Plotando um Horizontal
plt.figure(num = 2, figsize = (10,10))
plt.bar(meses,frequencia,color = "Blue")

plt.show


```
